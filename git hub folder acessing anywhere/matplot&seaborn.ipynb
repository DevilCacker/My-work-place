```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
a = np.array([1,2,3,4,5])
b = np.array([12,13,14,18,20])
```

```python
plt.title("ggggggg")#heading of graph
plt.xlabel("a")#x-axis name
plt.ylabel("vdc")#y-axis name
plt.plot(a,b)#its assigning the a,b
plt.show()#to print thr graph
```

```python
plt.title("ggggggg")#heading of graph
plt.xlabel("a")#x-axis name
plt.ylabel("vdc")#y-axis name
plt.plot(a,b,'bo')#its assigning the a,b,bo reprsenting the point only
plt.show()#to print thr graph
```

```python
plt.title("ggggggg")#heading of graph
plt.xlabel("a")#x-axis name
plt.ylabel("vdc")#y-axis name
plt.plot(a,b,color = 'k')#its assigning the a,b,color=to change line pf graph
plt.show()#to print thr graph
```

```python
plt.title("ggggggg")#heading of graph
plt.xlabel("a")#x-axis name
plt.ylabel("vdc")#y-axis name
plt.plot(a,b,color = 'k',marker='o')#its assigning the a,b,color, marker reprsenting the plotting models#line chart(plt.plot())
plt.show()#to print thr graph
```
